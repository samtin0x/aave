```python
flipside = Flipside(FLIPSIDE_API_KEY, "https://api-v2.flipsidecrypto.xyz")
```

```python
def get_flipside_data(sql):
    all_rows = []
    current_page_number = 1
    page_size = 15000
    
    query_result_set = flipside.query(sql, page_number=current_page_number, page_size=page_size)
    
    total_pages = query_result_set.page.totalPages

    while current_page_number <= total_pages:
        results = flipside.get_query_results(
        query_result_set.query_id,
        page_number=current_page_number,
        page_size=page_size
      )

        total_pages = results.page.totalPages
        if results.records:
            all_rows = all_rows + results.records

        current_page_number += 1
    
    return pd.DataFrame(all_rows, columns = query_result_set.columns) #all_rows, query_result_set.columns
```

5 . **Create market overview dataset**

1. **Action Segregation**: Transactions are categorized by type (deposit, withdraw, borrow, repay) at each unique block timestamp.
   
2. **Cumulative Calculations**: Cumulative sums provide a continuous view of market activities, including deposits, withdrawals, borrows, and repays.

3. **Borrowing Dynamics**: The total borrowed amount is calculated, key for assessing the borrowing landscape.

4. **Total Value Locked (TVL) Metrics**: Two forms of TVL are calculated, one including borrows and the other excluding. TVL counts the tokens locked in the contracts, used as collateral or to earn yield. Borrowed coins aren't counted towards the TVL, avoiding inflation through cycled lending. This measure reflects actual liquidity in the contracts, essential for our strategic decision-making.

The dataset, `all_blocks_v3`, provides a multidimensional view of AAVE v3

```python
# Separate the actions into their respective columns and group by 'evt_block_time'
all_blocks_v3 = aave_v3.groupby('evt_block_time').apply(
    lambda x: pd.Series({
        'deposits': x.loc[x['action'] == 'deposit', 'amount_usd'].sum(),
        'withdraws': x.loc[x['action'] == 'withdraw', 'amount_usd'].sum(),
        'borrows': x.loc[x['action'] == 'borrow', 'amount_usd'].sum(),
        'repays': x.loc[x['action'] == 'repay', 'amount_usd'].sum(),
    })
).fillna(0)
all_blocks_v3.sort_index(inplace=True)

# Calculate cumulative sums for actions
all_blocks_v3['cumulative_deposits'] = all_blocks_v3['deposits'].cumsum()
all_blocks_v3['cumulative_withdraws'] = all_blocks_v3['withdraws'].cumsum()
all_blocks_v3['cumulative_borrows'] = all_blocks_v3['borrows'].cumsum()
all_blocks_v3['cumulative_repays'] = all_blocks_v3['repays'].cumsum()

# Calculate Total Borrowed Amount and Total Value Locked
all_blocks_v3['total_borrowed'] = all_blocks_v3['cumulative_borrows'] - all_blocks_v3['cumulative_repays']
all_blocks_v3['total_value_locked_inc_borrows'] = all_blocks_v3['cumulative_deposits'] - all_blocks_v3['cumulative_withdraws']
all_blocks_v3['total_value_locked_exl_borrows'] = all_blocks_v3['total_value_locked_inc_borrows'] - all_blocks_v3['total_borrowed']
```

6 . **Assessing risk: LTV Ratio by user throught time**

In this section, we'll delve into the intricacies of individual user behavior to gauge their average Loan-to-Value (LTV) and risk appetite. Analyzing the actions that users commonly undertake, we can construct a nuanced understanding of how they interact with the platform.

Evaluating the LTV over time serves multiple critical functions in our analysis:
1. **Understanding Risk Behavior:** By tracking the LTV, we can categorize users based on their risk preferences, differentiating between conservative and aggressive strategies.
2. **Modeling Profit Contribution:** Understanding the users' behaviors helps in forecasting their likely future actions and their potential profit contribution to AAVE through fees.
3. **Tailoring Strategies:** This insight is vital for personalized marketing or product development, ensuring that AAVE's offerings align with the users' needs and risk profiles.

This comprehensive examination will allow us to create robust features about users' behavior


```python
aave_v3_ltv_historicals = aave_v3_cutoff.copy()

aave_v3_ltv_historicals['month'] = aave_v3_ltv_historicals['evt_block_time'].dt.to_period('M')
aave_v3_ltv_historicals.sort_values(['user', 'evt_block_number', 'evt_index'], inplace=True)

# Initialize columns for cumulative deposits, borrows, and LTV
aave_v3_ltv_historicals['cumulative_deposits'] = 0.0
aave_v3_ltv_historicals['cumulative_borrows'] = 0.0
aave_v3_ltv_historicals['ltv'] = 0.0

# Iterate through the DataFrame to calculate cumulative values and LTV
for user in aave_v3_ltv_historicals['user'].unique():
    user_data = aave_v3_ltv_historicals[aave_v3_ltv_historicals['user'] == user]

    cumulative_deposits = 0
    cumulative_borrows = 0

    for index, row in user_data.iterrows():
        action = row['action']
        amount = row['amount_usd']

        if action == 'deposit':
            cumulative_deposits += amount
        elif action == 'withdraw':
            cumulative_deposits -= amount
        elif action == 'borrow':
            cumulative_borrows += amount
        elif action == 'repay':
            cumulative_borrows -= amount
        elif action == 'liquidation':
            cumulative_deposits -= amount
            cumulative_borrows -= amount

        aave_v3_ltv_historicals.at[index, 'cumulative_deposits'] = cumulative_deposits
        aave_v3_ltv_historicals.at[index, 'cumulative_borrows'] = cumulative_borrows

        # Calculate LTV if there's collateral
        if cumulative_deposits > 0:
            aave_v3_ltv_historicals.at[index, 'ltv'] = cumulative_borrows / cumulative_deposits

        
aave_v3_ltv_historicals.head()
```

```python
aave_v3_ltv_historicals['month'] = aave_v3_ltv_historicals['evt_block_time'].dt.to_period('M')

aave_v3_ltv_historicals['action_order'] = aave_v3_ltv_historicals.groupby('user').cumcount() + 1

# Example:
aave_v3_ltv_historicals[ aave_v3_ltv_historicals.user == "0xdda85a45ad5437ff9de9a0521370883d12d1ad56"][['action_order','evt_block_time','ltv','action','amount_usd','symbol','cumulative_deposits','cumulative_borrows']]
```

9. **Define cohorts based on the first transaction month**

```python
# Define cohorts based on the first transaction month
aave_v3_ltv_historicals['cohort_ID'] = aave_v3_ltv_historicals.groupby('user')['evt_block_time'].transform('min').dt.to_period('M')
```

9. **Create user summary data**

```python
stable_coins = ['USDC', 'USDT', 'DAI', 'GHO']

def classify_categories(symbols):
    unique_symbols = set(symbols)
    if unique_symbols.issubset(stable_coins):
        return 'stable_coin_only'
    elif unique_symbols.isdisjoint(stable_coins):
        return 'non_stables_only'
    else:
        return 'all'

    
user_summary = aave_v3_ltv_historicals.groupby(['user', 'cohort_ID']).agg(
    first_transaction_date=('evt_block_time', 'min'),
    is_aave_v2_user=('user', lambda x: any(user in aave_v2_cutoff['user'].values for user in x)),

    total_transactions=('action_order', 'max'),
    total_usd_transacted=('amount_usd', 'sum'),
    avg_transaction_value=('amount_usd', 'mean'),

    total_usd_flashloans=('amount_usd', lambda x: x[aave_v3_ltv_historicals['action'] == 'flashloan'].sum()),
    
    number_of_deposits=('action', lambda x: (x == 'deposit').sum()),
    number_of_borrows=('action', lambda x: (x == 'borrow').sum()),
    number_of_repays=('action', lambda x: (x == 'repay').sum()),
    number_of_liquidations=('action', lambda x: (x == 'liquidation').sum()),
    number_of_withdraws=('action', lambda x: (x == 'withdraw').sum()),
    
    average_ltv=('ltv', lambda x: x[x > 0].mean()), # Including only borrowers
    average_ltv_borrow=('ltv', lambda x: x[aave_v3_ltv_historicals.loc[x.index, 'action'] == 'borrow'].mean()),

    number_of_unique_symbols_transacted=('symbol', 'nunique'),
    trading_assets_category_type=('symbol', lambda x: classify_categories(x.unique())),
    number_of_unique_months_active=('evt_block_time', lambda x: x.dt.to_period("M").nunique())
    
).reset_index()
```

- Does a liquidation event lead to churn?

1. It initializes a liquidated flag as False.
2. It iterates through the user actions in the order given by action_order.
3. If it encounters a 'liquidation' action, it sets the liquidated flag to True.
4. If the liquidated flag is True, and the next action is 'borrow', it returns True, meaning the user borrowed again after being liquidated.
5. If it doesn't find a 'borrow' action after a 'liquidation', it returns False.


```python
liquidated_users = aave_v3_ltv_historicals[aave_v3_ltv_historicals['action'] == 'liquidation']['user'].unique()

# Function to check if a user borrowed after being liquidated
def did_borrow_after_liquidation(user_actions):
    liquidated = False
    borrowed = 0
    for action, order, amount in zip(user_actions['action'], user_actions['action_order'], user_actions['amount_usd']):
        if liquidated and action == 'borrow':
            borrowed = borrowed + amount
            return True
        if action == 'liquidation':
            liquidated = True
    return False

# Check which users borrowed after being liquidated
borrowed_after_liquidation = aave_v3_ltv_historicals[aave_v3_ltv_historicals['user'].isin(liquidated_users)].groupby('user').apply(did_borrow_after_liquidation)

# Calculate the percentage of users who were liquidated and never borrowed again
churned_percentage = (len(liquidated_users) - borrowed_after_liquidation.sum()) / len(liquidated_users) * 100

print(f"{churned_percentage:.2f}% of liquidated users did not borrow again.")
```

```python
result_list = []

for user in common_users:
    user_data = aave_v3_ltv_historicals[aave_v3_ltv_historicals['user'] == user].sort_values(by='evt_block_time')
    
    # Extract deposit and borrow date
    deposit_date = user_data[user_data['action'] == 'deposit']['evt_block_time'].iloc[0]
    borrow_date = user_data[user_data['action'] == 'borrow']['evt_block_time'].iloc[0]
    
    # Calculate days difference
    days_difference = (borrow_date - deposit_date).days
    
    # Extract cohort ID
    cohort_id = user_data['cohort_ID'].iloc[0]
    
    # Append the result
    result_list.append({'user': user, 'days_difference': days_difference, 'cohort_ID': cohort_id})

average_days_df = pd.DataFrame(result_list)
average_days_by_cohort = average_days_df.groupby('cohort_ID')['days_difference'].mean().reset_index()
```

## Cohorts (ML)

```python
user_summary_temp = user_summary.copy()

user_summary_temp = user_summary_temp
user_summary_temp = pd.get_dummies(user_summary_temp, columns=['trading_assets_category_type'])
features = ['trading_assets_category_type_all', 'trading_assets_category_type_non_stables_only','trading_assets_category_type_stable_coin_only','total_transactions', 'total_usd_transacted', 'avg_transaction_value',
            'number_of_borrows', 'number_of_repays', 'number_of_liquidations', 'average_ltv_borrow']
```

```python
for feature in features:
    Q1 = user_summary_temp[feature].quantile(0.01)
    Q3 = user_summary_temp[feature].quantile(0.99)
    IQR = Q3 - Q1
    user_summary_temp = user_summary_temp[(user_summary_temp[feature] >= Q1) & (user_summary_temp[feature] <= Q3)]
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

NUMBER_CLUSTERS = 4


def human_format(num):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    return f"{num:.2f}{['', 'K', 'M', 'B', 'T'][magnitude]}"


X = user_summary_temp[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Build and train the K-means model
kmeans = KMeans(n_clusters=NUMBER_CLUSTERS, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels back to the DataFrame
user_summary_temp['cluster'] = clusters

# Analyze the clusters
for i in range(NUMBER_CLUSTERS):
    print(f"Cluster {i}:")
    means = user_summary_temp[user_summary_temp['cluster'] == i][features].mean()
    _sum = user_summary_temp[user_summary_temp['cluster'] == i][features].sum()
    for feature, mean, _sum in zip(features, means, _sum):
        print(f"{feature}: {human_format(mean)} mean; {human_format(_sum)} sum")
    print("-----------")
```

```python
clusters = [0, 1, 2, 3]

# Set colors for different clusters
colors = ['b', 'g', 'r', 'y']

# Loop through each cluster and plot
for i in clusters:
    subset = user_summary_temp[user_summary_temp['cluster'] == i]
    plt.scatter(subset['average_ltv'], subset['total_usd_transacted'], c=colors[i], label=f'Cluster {i}')

plt.xlabel('Average LTV')
plt.ylabel('Total USD Transacted')
plt.title('Scatter Plot of Average LTV vs Total USD Transacted')
plt.legend()
plt.show()
```
